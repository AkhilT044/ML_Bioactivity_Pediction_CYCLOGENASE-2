```python
import pandas as pd


df = pd.read_csv('ChEMBL221_14062025_raw_main_columns.csv').dropna()
```

```python
print("Available columns:", df.columns.tolist())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df = df[df[['standard_value', 'standard_type']].notna().all(axis=1)]
```

```python
df = df[df['standard_type'].str.upper() == "IC50"]
```

```python
bioactivity_class = [
    "inactive" if float(val) >= 10000 else "active"  
    for val in df.standard_value
]
```

```python
df['bioactivity_class'] = bioactivity_class
df
```

```python
print(df['bioactivity_class'].value_counts())
```

```python
df = df[['molecule_chembl_id', 'smiles', 'standard_value', 'bioactivity_class']]
```

```python
df.to_csv('ChEMBL221_14062025_preprocessed.csv', index=False)

print(" Preprocessing complete. Preprocessed file saved as 'ChEMBL221_14062025_preprocessed.csv'.")

df
```
